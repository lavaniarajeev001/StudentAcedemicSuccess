```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
import shap
from sklearn.linear_model import Lasso
```

```python
df = pd.read_csv("Predict Student Dropout and Academic Success.csv")
```

```python
df.head(1)
```

```python
df.isna().sum()
```

```python
df.Target.value_counts()
```

```python
df.Target=df.Target.map({"Graduate":1,"Dropout":2,"Enrolled":3})
```

```python
X = df[['Course', 'Previous qualification (grade)', "Father's occupation",
        'Admission grade', 'Age at enrollment',
        'Curricular units 1st sem (approved)',
        'Curricular units 1st sem (grade)',
        'Curricular units 2nd sem (evaluations)',
        'Curricular units 2nd sem (approved)',
        'Curricular units 2nd sem (grade)']]
y=df["Target"]
```

```python
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X_scaled,y,test_size=.30,random_state=42)
```

```python
#ANN
model = Sequential()
```

```python
X_train.shape
```

```python
model.add(Dense(64,activation="relu",input_dim=10))
model.add(Dropout(0.50))
model.add(Dense(32,activation="relu"))
model.add(Dropout(0.30))
model.add(Dense(1,activation="sigmoid",))

model.compile(optimizer="adam",loss="CategoricalCrossentropy",metrics=["accuracy"])

model.fit(X_train,y_train,batch_size=10,epochs=100,validation_split=0.20)
```
